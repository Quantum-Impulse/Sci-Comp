```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math 

# plot the function 1/1+x from -10 to 10
x = np.linspace(-10, 10, 100)


# obtain square root of using binary search, max torlerance of 0.00001
def sqrt(x):
    # edge cases
    if x < 0:
        return None
    if x == 0:
        return 0
    if x == 1:
        return 1
    
    # binary search
    left = 0
    right = x
    
    while left < right and right - left > 0.00001:
       
        mid = (left + right) / 2
        
        if mid * mid == x:
            return mid
        elif mid * mid < x:
            left = mid
        else:
            right = mid
    return left

# plot the intterations of the square root of 7
x = 7
y = [x]
for i in range(10):
    x = sqrt(x)
    y.append(x)

plt.plot(y, color='blue')
plt.title('Square Root of 7')
plt.xlabel('Iterations')
plt.ylabel('Value')
plt.show()

```

```python
# babylonian method for square root
def sqrt(x):
    # edge cases
    if x < 0:
        return None
    if x == 0:
        return 0
    if x == 1:
        return 1
    
    # babylonian method
    old_guess = x
    new_guess = x/2.0
   
    while abs(old_guess - new_guess) > 0.00001:
        old_guess = new_guess
        new_guess = (old_guess + x / old_guess) / 2.0
    
    return new_guess

# plot the intterations of the square root of 7
x = 7
y = [x]
for i in range(10):
    x = sqrt(x)
    y.append(x)

plt.plot(y, color='blue')
plt.title('Square Root of 7')
plt.xlabel('Iterations')
plt.ylabel('Value')
plt.show()
```

```python
# define function 
def fun(x):
    return np.sin(x) + x*x - 2 * np.log(x) - 5

# find the root of the function using binary search, bisect function
from scipy.optimize import bisect

root_scipy = bisect(fun, 1, 3)


# create our own bisect function, track the iterations and plot them
def bisect(fun, a, b, tol=0.00001):
    # edge case
    if fun(a) * fun(b) > 0:
        return None
    
    # initialize the list to store the iterations
    iterations = np.array([])
    
    # binary search
    while b - a > tol:
        mid = (a + b) / 2
        iterations = np.append(iterations, mid)
        
        if fun(mid) == 0:
            return mid
        elif fun(mid) * fun(a) < 0:
            b = mid
        else:
            a = mid
    return iterations


# find root and plot each iteration with our own bisect function
root = bisect(fun, 2, 3)

# plot the iterations, and zoom near the root
plt.plot(root, fun(root), 'ro',)
plt.title('Root of the Function')
plt.xlabel('Iterations')
plt.ylabel('Value')
plt.xlim(0, 10)
plt.show()



print("our root: ", root)
print("scipy root: ", root_scipy) 
```

```python
import numpy as np
# regula falsi method to get roots
def regula_falsi(fun, a, b, tol=0.00001):
    f_left = fun(a)
    f_right = fun(b)

    if f_left * f_right > 0:
        return None
    
    # initialize the list to store the iterations
    iterations = np.array([])

    while b - a > tol:
        # check if mid is zero
        if f_left == 0:
            return a
        if f_right == 0:
            return b
        
        mid = (a * f_right - b * f_left) / (f_right - f_left)

        iterations = np.append(iterations, mid)
        
        f_mid = fun(mid)
        if f_mid == 0:
            return mid
        elif f_mid * f_left < 0:
            b = mid
            f_right = f_mid
        else:
            a = mid
            f_left = f_mid

    return iterations

fun = lambda x: x*x - 2  
root = regula_falsi(fun, 0, 2)
print("Root: ", root.item(-1))


```

```python
# define function 
def fun(x):
    return x * x - 2
```

```python
# deffine the derivative of the function
def der(x):
    return 2 * x
```

```python
# define the Newton-Raphson method
def newton_raphson(fun, der, x, tol=0.00001):
    # initialize the list to store the iterations
    iterations = np.array([])
    
    # newton raphson
    while abs(fun(x)) > tol:
        iterations = np.append(iterations, x)
        x = x - fun(x) / der(x)
    return iterations

# find the root of the function
root = newton_raphson(fun, der, 2)

print("Root: ", root.item(-1))
```

```python
# Secant method
def secant(fun, x0, x1, tol=0.00001):
    # initialize the list to store the iterations
    iterations = np.array([])
    
    # secant method
    while abs(fun(x1)) > tol:
        iterations = np.append(iterations, x1)
        x2 = x1 - fun(x1) * (x1 - x0) / (fun(x1) - fun(x0))
        x0 = x1
        x1 = x2
    return iterations

# find the root of the function
root = secant(fun, 1, 2)
print("Root: ", root.item(-1))
```
