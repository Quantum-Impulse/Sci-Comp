```python
import sys

def main():
    line = sys.stdin.readline().strip()

    line = line.rstrip()

    line = line.split()

    num = int(line[0])
    base = int(line[1])

    s = ''
    while num > 0:
        s = str(num % base) + s
        num = num // base

    
    print(s)
main()
    

```

```python
# collatz sequence length function

def collatz(n):
    count = 1
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3*n + 1
        count += 1
    return count

print(collatz(7)) # 7)

# find the number that has the longest collatz sequence length
def long__collatz(n):
    longest = 0
    num = 0
    for i in range(1, n):
        if collatz(i) > longest:
            longest = collatz(i)
            num = i
    return num

print(long__collatz(100)) # 837799


```

```python
# create a fucntion, to the plot and find minimum 
# function (x-3)**2 -5

import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return (x - 3)**2 - 5
x = np.linspace(-10, 10, 100)
y = f(x)

plt.plot(x, y)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.title('Plot of f(x) = (x - 3)**2 - 5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()

# find the minimum of the function
from scipy.optimize import minimize_scalar
result = minimize_scalar(f)
print(f"The minimum value of the function is at x = {result.x:.2f}, f(x) = {result.fun:.2f}")
# The minimum value of the function is at x = 3.00, f(x) = -5.00


# find the roots of the function
from scipy.optimize import fsolve
def f(x):
    return (x - 3)**2 - 5
root = fsolve(f, 0)
print(f"The root of the function is at x = {root[0]:.2f}")
# The root of the function is at x = 3.00



```
