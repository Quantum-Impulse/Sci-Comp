```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
mu_min, mu_max = 2.4, 4.0  # Range of mu values
num_mu = 1000  # Number of mu values
num_iterations = 1000  # Total iterations per mu
transient_iterations = 200  # Discard these to remove transient effects
x0 = 0.5  # Initial value of x

# Create an array of mu values
mu_values = np.linspace(mu_min, mu_max, num_mu)

# Prepare figure
plt.figure(figsize=(10, 6), dpi=300)

# Iterate over each mu value
for mu in mu_values:
    x = x0  # Reset initial condition for each mu
    
    # Iterate the logistic map
    for i in range(num_iterations):
        x = mu * x * (1 - x)  # Logistic map equation
        
        # After transient iterations, collect points for bifurcation diagram
        if i >= transient_iterations:
            plt.plot(mu, x, ',k', alpha=0.2)  # Small black dots

# Configure plot
plt.title("Bifurcation Diagram of the Logistic Map")
plt.xlabel("Mu ()")
plt.ylabel("x values")
plt.xlim(mu_min, mu_max)
plt.ylim(0, 1)
plt.show()
```
