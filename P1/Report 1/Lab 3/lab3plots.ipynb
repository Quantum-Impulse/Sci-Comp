```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: frequencies (Hz) and corresponding gains (dB)
frequencies = np.linspace(1000, 30000, 13)  # from 20 Hz to 20 kHz
v_out = [98, 130, 144, 148, 148, 150, 152, 152, 152, 154, 148, 144, 136] 
v_in = 1.06
gain = [v / v_in for v in v_out]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(frequencies, gain, label='Gain (dB)', color='purple')
plt.xscale('log')  # logarithmic scale for frequency
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (dB)')
plt.title('Gain vs Frequency')
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='r')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given or measured values
L = 1e-3  # Inductance in Henrys (example value)
C = 8.18e-6  # Capacitance in Farads (example value)
R = 500  # Resistance in Ohms (example value)

# Resonant frequency (ω0 = 1 / sqrt(LC)) and convert to Hz
omega0 = 1 / np.sqrt(L * C)
frequency0 = omega0 / (2 * np.pi)

# Frequency range for measurement
frequencies = np.linspace(frequency0 * 0.8, frequency0 * 1.2, 12)  # 12 data points

# Calculate the gain as a function of frequency
omega = 2 * np.pi * frequencies
gain = (1 / R) / np.sqrt((1/(L*C)) - omega**2)**2 + (R / L)**2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(frequencies, gain, label='Gain vs Frequency', color='blue')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (V_out / V_in)')
plt.title('Gain vs Frequency for Bandpass LRC Circuit')
plt.grid(True)
plt.legend()

# Find the peak gain
peak_gain = np.max(gain)
peak_index = np.argmax(gain)
fwhm_indices = np.where(gain > peak_gain / 2)[0]

# Calculate FWHM and Q
if len(fwhm_indices) >= 2:
    fwhm = frequencies[fwhm_indices[-1]] - frequencies[fwhm_indices[0]]
else:
    fwhm = 0  # In case FWHM calculation fails

Q = frequency0 / fwhm if fwhm != 0 else np.inf

# Show plot
plt.show()

# Output Q factor
print(f"Resonant Frequency (Hz): {frequency0:.2f}")
print(f"Full Width at Half Max (Hz): {fwhm:.2f}")
print(f"Q Factor: {Q:.2f}")
```
