```python
import numpy as np
import matplotlib.pyplot as plt

# Load resistor values (in ohms)
R_load = np.array([100, 220, 330, 470, 680, 1000, 2200, 3300, 4700, 10000])

# Corresponding Vout values (in volts)
V_out = np.array([1.5, 2.2, 3.0, 3.8, 4.2, 4.5, 4.7, 4.8, 4.9, 4.95])

# Calculate 1/R_load and 1/V_out
inv_R_load = 1 / R_load
inv_V_out = 1 / V_out

# Plot
plt.figure(figsize=(6,4))
plt.plot(inv_R_load, inv_V_out, 'bo-', label='Measured Data')
plt.xlabel(r'$1/R_{\text{Load}}$ ($\Omega^{-1}$)')
plt.ylabel(r'$1/V_{\text{out}}$ (V$^{-1}$)')
plt.title(r'Plot of $1/V_{\text{out}}$ vs $1/R_{\text{Load}}$')
plt.grid(True)
plt.legend()
plt.savefig('voltage_divider_plot.png', dpi=300)  # Save the plot
plt.show()
```
