```python
# import libs
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Basic plot
x = np.linspace(0, 10, 20)

y = x ** 2

plt.plot(x, y, color='blue', linewidth=2, linestyle='--')
plt.title('Numbers Squared')
plt.xlabel('X Axis: Numbers')
plt.ylabel('Y Axis: Numbers squared') # Labeling the a

plt.show()
```

```python
# plot cos x through -2pi to +3pi, have 100 points, color red
x = np.linspace(-3 * np.pi, 3 * np.pi, 100)
y = np.cos(x)

plt.plot(x, y, color='red')
plt.title('Cosine of x')
plt.xlabel('X Axis: Numbers')
```

```python
# plot taylor series of cos x , first 4 terms, color green
x = np.linspace(-3 * np.pi, 3 * np.pi, 100)

y = 1 - x**2/2 + x**4/24 - x**6/720 +  x**8/40320

plt.plot(x, y, color='green')
plt.title('Taylor Series of Cosine')
plt.xlabel('X Axis: Numbers')
```
